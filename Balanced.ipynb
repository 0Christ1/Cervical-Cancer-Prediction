```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00383/risk_factors_cervical_cancer.csv',na_values='?')
pd.set_option('display.max_rows', None, 'display.max_columns', None)
df.head(10)
```

```python

df['Biopsy'] =df['Biopsy']+df['Citology']+df['Schiller']+df['Hinselmann']

df.rename(columns={'Biopsy':'result'}, inplace=True)
```

```python
df.isnull().sum()
```

```python
df= df.drop(['STDs: Time since first diagnosis','STDs: Time since last diagnosis'],axis=1)
```

```python
df= df.drop(['Hinselmann', 'Schiller', 'Citology'],axis=1)
df = df.fillna(df.mean())
df_2=df

df_2.head(10)



```

```python
df_1 = df_2.copy()
df_1['result']=df_1['result'].astype(bool)   
df_1.head(10)
```

```python
df_2.head(10)
```

```python

from sklearn import preprocessing # preprossing is what we do with the data before we run the learning algorithm
from sklearn.model_selection import train_test_split 
import math

import matplotlib.pyplot as plt

y = df_1['result']
X= df_1.drop(['result'], axis=1)
# X = df3.iloc[:, 0:-4]
# y = df3.iloc[:, -1]

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=1)

# X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=1) # 0.25 x 0.8 = 0.2


from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
X_train, y_train = sm.fit_resample(X_train, y_train)

scaler = StandardScaler()
X_train= scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)


print("direction1")
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))

print("X_val shape: {}".format(X_val.shape))
print("y_val shape: {}".format(y_val.shape))


```

```python
import sklearn as sk
from sklearn.linear_model import LogisticRegression
import pandas as pd
import os


model = LogisticRegression()
model.fit(X_train, y_train)

result = model.score(X_train, y_train)
print("Logistic Regression, No Regularization, degree 1")
print("training accuracy = "+"{:.6f}".format(result))

# result = model.score(X_test, y_test)
# print("{:.6f}".format(result))

result = model.score(X_val, y_val)
print("validation accuracy = "+"{:.6f}".format(result))

# yhat = model.predict(X_val)
# acc = np.mean(yhat == y_val)
# print("Accuracy on test data = %f" % acc)
```
